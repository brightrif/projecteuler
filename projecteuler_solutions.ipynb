# Multiples of 3 and 5

Problem 1

If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.


# solution 
The number starts from 3 all the way to 999
muliples can be find through mod operator 


```python
x = 1

sum = 0
while x<1000:
    if x%3==0 or x%5==0:
        sum = sum+x
    x = x+1
print(sum) 
```

# Largest prime factor
Problem 3

The prime factors of 13195 are 5, 7, 13 and 29.

What is the largest prime factor of the number 600851475143 ?


# solution

```python
n = int(input('The largest prime factor of')) 

p = 2
while (p*p <= n):      
  if (n % p == 0): 
    n //= p
  else: 
    p += 2 if p>2 else 1   # after 2, consider only odd p

print(n)
```

```python
import math


def prime_factors(num):
    """
    Returns a list of primes factors for the given number.
    """
    factors = []
    # Handle 2
    while num % 2 == 0:
        factors.append(2)
        num /= 2

    # Handle all odd nums <= sqrt(num)
    for i in range(3, int(math.sqrt(num)) + 1, 2):
        while num % i == 0:
            factors.append(i)
            num /= i

    # Handle num is prime
    if num > 2:
        factors.append(num)

    return factors

prime_factors(600851475143)
```
